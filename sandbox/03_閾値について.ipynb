```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([0.41569836, 1.5470663, 0.62904596, 0.4805466, 0.61101004, 0.37301317,
                 0.51961979, 0.63458617, 0.4234963, 0.39070821, 0.48245548, 0.4713493,
                 0.4769712, 0.82421088, 0.43562191, 0.54986596, 0.47102947, 0.48542146,
                 0.46980413, 0.39323007, 0.44394635, 0.59839574, 0.72217701, 0.39859159,
                 0.39518409, 0.40250446, 0.74670511, 1.74752068, 0.94670939, 0.61781923])
```

```python
# 方法1: 標準偏差を用いた方法
mean = np.mean(data)
std_dev = np.std(data)
threshold_std_dev = mean + 3 * std_dev

# 方法2: 分位数を用いた方法
threshold_percentile_95 = np.percentile(data, 95)
threshold_percentile_99 = np.percentile(data, 99)

# 方法3: IQRを用いた方法
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
threshold_IQR = Q3 + 1.5 * IQR

# 方法4: 最大値を用いた方法
threshold_max = np.max(data)

# 方法5: 正規分布に基づく統計的手法
z_score = 1.96
threshold_sta = mean + z_score * std_dev

# 各閾値を表示
print("標準偏差を用いた閾値:", threshold_std_dev)
print("分位数を用いた閾値(95%):", threshold_percentile_95)
print("IQRを用いた閾値:", threshold_IQR)
print("最大値を用いた閾値:", threshold_max)
print("正規分布を用いた閾値:", threshold_sta)

_,ax = plt.subplots()
ax.plot(data,np.zeros(len(data)),'o')
ax.vlines(threshold_std_dev,-1,1,'r',label='std')
ax.vlines(threshold_percentile_95,-1,1,'g',label='percentile(95%)')
ax.vlines(threshold_percentile_99,-1,1,'m',label='percentile(99%)')
ax.vlines(threshold_IQR,-1,1,'b',label='iqr')
ax.vlines(threshold_max,-1,1,'c',label='max')
ax.vlines(threshold_sta,-1,1,'b',label='sta')

ax.set(xlabel='anomaly',ylabel='',yticklabels='',xlim=[0,None])
ax.legend(loc='upper left')
```
