```python
class Trigger:
    def __init__(self,model) -> None:
        self.model = model
        self.is_active = False

    def start(self):
        self.is_active = True

    def stop(self):
        self.is_active = False

class Model:
    def __init__(self) -> None:
        self._config = {}
        self._trigger = None

    @property
    def config(self):
        return self._config
    
    @config.setter
    def config(self,value):
        self._config = value
        self.trigger = Trigger(self)

    @property
    def trigger(self):
        return self._trigger
    
    @trigger.setter
    def trigger(self,value):
        self._trigger = value


class Controller:
    def __init__(self,model) -> None:
        self.model = model

model = Model()
model.config = None
print(model.trigger.is_active)

controller = Controller(model)
model.config = None
controller.model.trigger.start()
print(model.trigger.is_active)
```
